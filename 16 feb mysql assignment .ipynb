```python
 Q1. What is a database? Differentiate between SQL and NoSQL databases.
Ans : A database is an organized collection of structured information, or data, typically stored electronically in a computer system.
A database is usually controlled by a database management system (DBMS).

                   SQL	                                                                        NoSQL

1.RELATIONAL DATABASE MANAGEMENT SYSTEM (RDBMS)	                               Non-relational or distributed database system.
2.These databases have fixed or static or predefined schema	                   They have dynamic schema
3.These databases are not suited for hierarchical data storage.	               These databases are best suited for hierarchical data storage.
4.These databases are best suited for complex queries	                       These databases are not so good for complex queries
5.Vertically Scalable	                                                       Horizontally scalable
6.Follows ACID property	                                                       Follows CAP(consistency, availability, partition tolerance)
7.Examples: MySQL, PostgreSQL, Oracle, MS-SQL Server, etc	                   Examples: MongoDB, GraphQL, HBase, Neo4j, Cassandra, etc



Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.
Ans : Data definition language (DDL) refers to the set of SQL commands that can create and manipulate the structures of a database. 
      DDL statements are used to create, change, and remove objects including indexes, triggers, tables, and views. 
      Common DDL statements include:
      create : it generates the new table ex. CREATE TABLE Artists (artistName, varchar);
      Drope :   removes the database from the table ex. ALTER TABLE Artists
                                                        DROP CONSTRAINT ArtistID;
      Alter : alter the table ex : ALTER TABLE Artists ADD PRIMARY KEY (artist_pk);

Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.
Ans : DML is an abbreviation of Data Manipulation Language.
      The DML commands in Structured Query Language change the data present in the SQL database. We can easily access, store, modify, update and delete the existing records
      from the database using DML commands.
      Insert : INSERT is most important data manipulation command in Structured Query Language, which allows users to insert data in database tables.
            eg. INSERT INTO Student (Stu_id, Stu_Name, Stu_Marks, Stu_Age) VALUES (104, Anmol, 89, 19);  
      Update : UPDATE is another most important data manipulation command in Structured Query Language, which allows users to update or modify the existing data in database tables.
            eg. UPDATE Product SET Product_Price = 80 WHERE Product_Id = 'P102' ; 
      Delete : DELETE is a DML command which allows SQL users to remove single or multiple existing records from the database tables.
            eg : DELETE FROM Product WHERE Product_Id = 'P202' ;  

Q4. What is DQL? Explain SELECT with an example.
Ans : DQL is used to fetch the data from the database.
      It uses only one command:
      SELECT : It is used to select the attribute based on the condition described by WHERE clause.
           eg : SELECT emp_name  
                FROM employee  
                WHERE age > 20;  

Q5. Explain Primary Key and Foreign Key.
Ans : Primary keys :
          A primary key is a column or a set of columns in a table whose values uniquely identify a row in the table. A relational database is designed to enforce 
          the uniqueness of primary keys by allowing only one row with a given primary key value in a table.
      Foreign keys :
          A foreign key is a column or a set of columns in a table whose values correspond to the values of the primary key in another table. 
          In order to add a row with a given foreign key value, there must exist a row in the related table with the same primary key value.

Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.
Ans : import mysql.connector
       mydb = mysql.connector.connect(
         host="localhost",
         user="abc",
         password="password"
       )
       mycursor = mydb.cursor()

     cursor() :It is an object that is used to make the connection for executing SQL queries. It acts as middleware between SQLite database connection and SQL query. 
               It is created after giving connection to SQLite database.
    Execute() : It helps to execute the output of the link sql connection .
     

Q7. Give the order of execution of SQL clauses in an SQL query.
Ans : FROM/JOIN: The FROM and/or JOIN clauses are executed first to determine the data of interest.
      WHERE: The WHERE clause is executed to filter out records that do not meet the constraints.
      GROUP BY: The GROUP BY clause is executed to group the data based on the values in one or more columns.
```
